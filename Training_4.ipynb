# ETL (Extract, Transform, Load)

## Extract

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/International_wheat_production_statistics'
```

```python
df=pd.read_html(url)[0]
```

```python
df.head()
```

## Transform

```python
df.info()
```

```python
df['Country']=df['Country'].astype('string')
```

```python
df['2019[1]']=df['2019[1]'].str.replace('[7]', '')
```

```python
df['2019[1]']= df['2019[1]'].astype('float')
```

```python
df['2019[1]'].unique()
```

```python
df['2019[1]'].str.replace('[7]', '')
```

```python
df.describe().T
```

```python
df.replace(to_replace={'[7]':''})
```

```python

cols_df=list(df.select_dtypes(include='object').columns)
for i in cols_df:
    df[i].replace('[7]', '')
```

```python
# Select object columns
object_columns = df.select_dtypes(include='object').columns

# Perform replacement for each object column
for col in object_columns:
    df[col] = df[col].str.replace(r'\[7\]', '',regex=True)
```

```python
# Select object columns
object_columns = df.select_dtypes(include='object').columns

# Perform replacement for each object column
for col in object_columns:
    df[col] = df[col].str.replace(r'\[8\]', '',regex=True)
```

```python
# Select object columns
object_columns = df.select_dtypes(include='object').columns

# Perform replacement for each object column
for col in object_columns:
    df[col] = df[col].str.replace(r'\[\d\]', '',regex=True)
```

```python
for i in object_columns:
    df[i] = df[i].astype('float')
```

```python
df.columns=df.columns.str.replace(r'\[\d\]', '',regex=True)
```

```python
df.info()
```
